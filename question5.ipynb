```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the x range
x = np.linspace(0, 1000, 100)   # 100 points between 0 and 1000
q = 1.6e-19       # charge in coulombs
v0 = 1.0          # example amplitude 
c = 3e8           # speed of light in m/s
# Defining the function y which is radiation reaction force
y = (2* q**2 * v0 * np.exp(x/100)*((3/5)*np.cos(30*x)-899.9999*np.sin(30*x)))/(3*c**3) # you can replace this with any function, e.g., x**2, np.exp(x), etc.

# Create the plot
plt.plot(x, y,label='F_R vs t', color='blue')

# Add labels and title
plt.xlabel('t')
plt.ylabel('F_R')
plt.title('Plot of Radiaction Reaction Force F_R as a function of time t')
plt.legend()

# Show grid and display the plot
plt.grid(True)
plt.show()
```
