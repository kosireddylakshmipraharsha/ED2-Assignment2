```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.626e-34 #Plank's constant
m = 9.1e-31   #Mass of electron
k = 1.38e-23  #Boltzman's constant
T_values=[300, 10**5] #Temperature

ne = 1        #Number density of electron
ni = 1        #Number density of ions
z = 1          #Atomic number

# Frequency range (Hz)
x = np.logspace(10, 15, 300)   # creates evenly spaced numbers between 10^10 to 10^15 HZ (for frequencies less than 10^10 the value of the function y remains constant as the value u is small)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))   
for i, T in enumerate(T_values):
    # Define Function
    A = h * x / (k * T)
    B = (1 + A)   #for small values if u, B=1.
    y = 1.0438* 10**(-20) * ne * ni * z**2 * m**(-2.5) * k**0.5 * T**0.5 * np.exp(-A) * ((2*m/(k*T))*B + B**2 + 1)
    
    # plot
    #We used log scale for both X an Y axis
    axes[i].loglog(x, y, label='Power per frequency', color='blue')
    axes[i].set_title(f'Bremsstrahlung Power Spectrum at {T} K ')
    axes[i].set_xlabel("Frequency (Hz)")
    axes[i].set_ylabel('Power emitted per unit frequency')
    axes[i].legend()
    axes[i].grid(True, which="both", ls="--")

plt.tight_layout()
plt.show()

```
